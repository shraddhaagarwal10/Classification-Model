#**SONAR: Rock VS Mine Prediciton**


## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Data Collection and Analysis

```python
sonar=pd.read_csv("Sonar Data.csv", header=None)
```

```python
sonar.head()
```

```python
sonar.describe()
```

```python
sonar.shape
```

## Seperating data and target

```python
X=sonar.drop(columns=60, axis=1)
y=sonar[60]
```

```python
print(X.shape)
print(y.shape)
```

## Train and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

## Model 1: Logistic Regression

### Model fitting

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(X_train, y_train)
```

### Evaluating Model on Training Set

```python
from sklearn.metrics import accuracy_score
#accuracy on train data
X_train_prediction=model1.predict(X_train)
train_set_accuracy_1=accuracy_score(X_train_prediction, y_train)
train_set_accuracy_1
```

### Evaluating Model on Test Set

```python
#accuracy on test data
X_test_prediction=model1.predict(X_test)
test_set_accuracy_1=accuracy_score(X_test_prediction, y_test)
test_set_accuracy_1
```

## Model 2: K Neighbors Classifier

### Model fitting

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(X_train, y_train)
```

### Evaluating Model on Training Set

```python
#accuracy on train data
X_train_prediction=model2.predict(X_train)
train_set_accuracy_2=accuracy_score(X_train_prediction, y_train)
train_set_accuracy_2
```

### Evaluating Model on Test Set

```python
#accuracy on test data
X_test_prediction=model2.predict(X_test)
test_set_accuracy_2=accuracy_score(X_test_prediction, y_test)
test_set_accuracy_2
```

## Model 3: Decision Tree Classifier

## Model fitting

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(X_train, y_train)
```

### Evaluating Model on Training Set

```python
#accuracy on train data
X_train_prediction=model3.predict(X_train)
train_set_accuracy_3=accuracy_score(X_train_prediction, y_train)
train_set_accuracy_3
```

### Evaluating Model on Training Set

```python
X_test_prediction=model3.predict(X_test)
test_set_accuracy_3=accuracy_score(X_test_prediction, y_test)
test_set_accuracy_3
```

## Accuracy of Test Set using above models

```python
print('Accuracy on Test Set using Model 1:', test_set_accuracy_1)
print('Accuracy on Test Set using Model 2:', test_set_accuracy_2)
print('Accuracy on Test Set using Model 3:', test_set_accuracy_3)
```

Since, Decision Tree Classifier(Model 3) overfits the training set so, it has worst prediction of test set among the three classifiers.
K Neighbors Classifier(Model 2) has the best prediction of test set among the three classifiers.
So, we will use Model 2 for our predicitive system.

## Making a Predictive System

```python
input_data=[float(s) for s in (input('Enter 60 features: ')).split()]
input_data_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_numpy_array.reshape(1,-1)
prediction=model2.predict(input_data_reshaped)
print(prediction)

if prediction[0]=='R':
    print('The object is a rock')
else:
    print('The object is a mine')
```
